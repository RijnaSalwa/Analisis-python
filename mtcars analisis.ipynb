```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
rcParams['figure.figsize']=15,12
from matplotlib.ticker import FuncFormatter
```

```python
mtcars=pd.read_excel('mtcars.xls', index_col=0)
```

```python
mtcars.head()
```

```python
mpg_level = []
for value in mtcars["mpg"]: 
    if value <= 20: 
        mpg_level.append("Low")  
    elif value >= 30: 
        mpg_level.append("hard") 
    else: 
        mpg_level.append("medium")
         
mtcars.insert(1, "mpg_level", mpg_level)
```

```python
mtcars.head()
```

```python
print (mtcars.describe())
```

```python
mtcars.plot(kind='box', subplots=True, layout=(4,3), sharex=False, sharey=False)
plt.show()
```

```python
mtcars.hist()
plt.show()
```

```python
scatter_matrix(mtcars)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(mtcars.index, mtcars.mpg)
```
